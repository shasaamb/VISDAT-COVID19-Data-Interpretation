# Data Preprocessing

## Data Population

```python
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/3. DATASET/EstimasiPopulasi.xlsx'
df_population = pd.read_excel(file_path)
df_population.head()
```

```python
df_population.columns = df_population.iloc[0]
df_population = df_population.drop(0)
df_population.head()
```

```python
df_population = df_population.dropna(axis=1, how='all')
columns_to_drop = ['']
df_population.head()
```

```python
columns_to_keep = ['Region, subregion, country or area *', 'Year', '0-4',	'5-14',	 '15-19',	'20-24',	 '25-39', '40-49',	'50-59',	 '60-69', 	'70-79',	 '80-89',	'90-99',	'100+']
df_population = df_population.loc[:, columns_to_keep]
df_population = df_population[columns_to_keep]
df_population.head()
```

## Data Death

```python
file_path = '/content/drive/MyDrive/3. DATASET/Kematian.xlsx'
df_death = pd.read_excel(file_path)
df_death.head()
```

```python
df_death.columns = df_death.iloc[0]
df_death = df_death.drop(0)
df_death.head()
```

```python
df_death = df_death.dropna(axis=1, how='all')
df_death.head()
```

```python
columns_to_keep = ['Region, subregion, country or area *', 'Year', '0-4',	'5-14',	 '15-19',	'20-24',	 '25-39', '40-49',	'50-59',	 '60-69', 	'70-79',	 '80-89',	'90-99',	'100+']
df_death = df_death.loc[:, columns_to_keep]
df_death = df_death[columns_to_keep]
#df_death = df_death.drop(df_death.columns[4], axis=1)
df_death.head()
```

```python
df_death = df_death.loc[:, ~df_death.columns.duplicated()]
df_death.head()
```

#EDA

## Mortality Rate

```python
df_mortality_rates = pd.DataFrame()

for year in range(1950, 2021):
    mortality_rate_year = df_death[df_death['Year'] == year].iloc[:, 2:] / df_population[df_population['Year'] == year].iloc[:, 2:] * 1000
    mortality_rate_year.insert(1, 'Year', year)

    df_mortality_rates = pd.concat([df_mortality_rates, mortality_rate_year], ignore_index=True)

column_order = ['Year'] + list(df_death.columns[2:])
df_mortality_rates = df_mortality_rates[column_order]
df_mortality_rates.head()
```

```python
descriptive_stats_year = df_mortality_rates.groupby('Year').describe()
descriptive_stats_age_group = df_mortality_rates.drop('Year', axis=1).describe()

print("Descriptive Statistics for Each Year:")
print(descriptive_stats_year)

print("\nDescriptive Statistics for Each Age Group:")
print(descriptive_stats_age_group)
```

# Data Visualization

```python
# Line Plot #
import matplotlib.pyplot as plt

df_mortality_rates.set_index('Year', inplace=True)

plt.figure(figsize=(10, 6))
for column in df_mortality_rates.columns:
    plt.plot(df_mortality_rates.index, df_mortality_rates[column], label=column)

plt.title('Mortality Rates Over Time for Each Age Group')
plt.xlabel('Year')
plt.ylabel('Mortality Rate')
plt.legend()
plt.grid(True)
plt.show()

###

from google.colab import files
plt.savefig("linePlot.png")
files.download("linePlot.png")
```

```python
df_aggregated = df_mortality_rates.groupby('Year').mean()

correlation = df_aggregated.corr().iloc[0, 1]
print("Correlation between Year and Mortality Rate:", correlation)
```

```python
# Comparative Analysis #
import seaborn as sns

df_mortality_rates = df_mortality_rates.apply(pd.to_numeric, errors='coerce')

df_mortality_rates.dropna(inplace=True)

plt.figure(figsize=(12, 8))
sns.heatmap(df_mortality_rates, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title('Comparative Analysis of Mortality Rates Across Different Age Groups and Years')
plt.xlabel('Age Group')
plt.ylabel('Year')
plt.show()

###

from google.colab import files
plt.savefig("compareAnalysis.png")
files.download("compareAnalysis.png")
```

```python
# Histogram #
plt.figure(figsize=(10, 6))
for column in df_mortality_rates.columns[1:]:
    sns.histplot(df_mortality_rates[column], kde=True, label=column, alpha=0.5)
plt.title('Distribution of Mortality Rate by Age Group')
plt.xlabel('Mortality Rate')
plt.ylabel('Frequency')
plt.legend()
plt.show()

###

from google.colab import files
plt.savefig("histogram.png")
files.download("histogram.png")
```

```python
# Box Plot #
df_mortality_rates_without_year = df_mortality_rates.drop(columns=['Year'])

plt.figure(figsize=(10, 6))
sns.boxplot(data=df_mortality_rates_without_year)
plt.title('Mortality Rate Distribution for Each Age Group')
plt.xlabel('Age Group')
plt.ylabel('Mortality Rate')
plt.show()

###

from google.colab import files
plt.savefig("boxPlot.png")
files.download("boxPlot.png")
```
